```python
# Here are the necessary libraries (extra code) that comes with downloading anaconda
from matplotlib import pyplot as plt
import numpy as np

# This makes all of the graphs show in this notebook
%matplotlib inline
```

## Below is a graph in python

Blue dots are the smart kids

Red x's are kids who aren't

This is a made-up graph, to show how machines learn to draw lines.

<br>

Let's say I had all the data from kids at Fisher about how much time they spent on instagram and how much time they spent doing homework.

If I want to learn what makes kids smart, I need to figure out what separates the smart kids from the not smart kids.


```python
def genericPlot():
    x = np.arange(0,5,1) # list of numbers from 0-5 spaced by 1
    y = [8,5,2,3,1]
    # Together (x,y) form a LIST of ordered pairs
    plt.scatter(x,y) # kids who play sports are blue dots

    x2 = np.arange(5,10) 
    y2 = [7,6,8,4,7] 
    # Together (x2,y2) form a LIST of ordered pairs
    plt.scatter(x2,y2,c='r',marker='x') # kids who don't play sports are red x's

    plt.ylim(0,10)
    plt.xlim(0,10)
    plt.xlabel("Half-Hours Spent Doing Homework")
    plt.ylabel("Half-Hours Spent On Instagram")
    plt.title("Smart Kids & Non-Smart Kids")
    graph = plt
    return graph

genericPlot().show()
```

# It turns out I can separate the kids with a line!
---

It turns out that there's a line with the equation

     y = -x + 9
     
     or 
     
     y = 9 - x
     
I can use this to write into a function in Python!

```python
def separateSmarties(x):
    return 9 - x
```

Now, If I plot this, I should be able to see the line separates the kids pretty well

```python
graph = genericPlot()
x = arange(0,10,0.1)
graph.plot(x,separateSmarties(x))
graph.show()
```

# What does this mean?

This line shows that if a kid is below the line, then he's probably not smart, and if he's above, he probably is smart.

It turns out that pretty much 90% of artificial intelligence works similar to this. All decisions in artificial intelligence are based on data, and drawing these lines in data is called machine learning. It's what lets AI become so smart; the more mistakes a machine makes, the more data it has to predict with.


```python
def newData():
    graph = genericPlot()

    # it may turn out that smart kids defy boundaries, and we had bad data

    newDataX = arange(2,6,1)
    newDataY = [8,6,5,4]
    graph.scatter(newDataX, newDataY, color='r',marker='x') # smarties

    newDataX = arange(8,11,1)
    newDataY = [1,3,6]
    graph.scatter(newDataX,newDataY)
    # just gonna throw in an extra point
    graph.scatter(2,7)
    return graph

newData().show()
```

# Oops!
---

If with this new data, if we drew our straight line again, we'd ***misclassify*** some non-smart kids as smart kids, and vice versa

```python
x = linspace(0,10,100)
graph = newData()
graph.plot(x,separateSmarties(x))
graph.show()
```

It turns out, we need better models to separate smart kids

```python
# Flexible models are important to machine learning, and those form the basis of AI
x = linspace(0,10,100)
graph = newData()
graph.plot(x,0.45*(x-6)**2)
graph.show()
```

These sorts of **new data problems** are pretty big in AI, and that's why it can be a difficult field. But also very interesting.

It turns out by studying the brain, we can come up with really good code to draw better lines. The algorithm is even named after the brain -- it's called a neural network! A new tutorial will hopefully be up soon!
